```python
FILE_PATH = "data/popular-names.txt"

# textを各行読み込み
with open(FILE_PATH, "r") as f:
    text_list = f.readlines()


name_set = set()    # 名前一覧
name_list = []      # 名前(重複あり)一覧
name_piece_list = []# [名前,出現回数]のリスト

# 名前の取り出し
for i in text_list:
    name,_ ,_ ,_= i.split("\t")
    name_set.add(name)
    name_list.append(name)

#名前の出現回数のカウント
for i in name_set:
    pieces = name_list.count(i)
    name_piece_list.append([i, pieces])

# name_piece_listを出現回数の降順に並べ替える
name_piece_list.sort(reverse = True, key = lambda x: x[1])

# name_piece_listから名前のみを取り出す
answer = list(map(lambda x: x[0], name_piece_list))

# len(answer) = 136なのでok
answer

#P.10 明確な単語を選ぶ
```
