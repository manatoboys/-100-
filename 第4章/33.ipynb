```python
#問題30の処理を関数化した
def prepare():   
    with open("neko.txt.mecab", "r") as f:
        data = f.readlines()    #各行の読み込み
        data = [line.strip("\n") for line in data]  #全ての行から"\n"を排除

    ans = []    #回答保存用配列

    #各行をループ
    for text in data:
        if text == "EOS":
            break
        dict = {}

        #空白の場合の処理
        if text[0] == " ":
            dict["surface"] = " "
            dict["base"] = " "
            temp = text.split("\t")
            print(temp)
            dict["pos"] =temp[0]
            dict["pos1"] = temp[1]
        #それ以外
        else:
            temp = text.split("\t")
            dict["surface"] = temp[0]
            dict["base"] = temp[1]
            dict["pos"] = temp[2]
            dict["pos1"] = temp[3]
        ans.append(dict)
    return ans
```

```python
data = prepare()
ans = []    #回答保存用

#3つずつ見ていく
for i in range(0,len(data)-3):
    text1 = data[i]
    text2 = data[i+1]
    text3 = data[i+2]
    # "の"で連結された名詞の特定
    if text2["surface"] == "の" and text1["pos"] == "名詞" and text3["pos"] == "名詞":
        str = text1["surface"] + text2["surface"] + text3["surface"] #"の"で連結された名詞
        ans.append(str)     #ansに格納

ans
```
