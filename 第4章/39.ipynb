```python
#問題30の処理を関数化した
def prepare():   
    with open("neko.txt.mecab", "r", encoding="utf-8") as f:
        data = f.readlines()    #各行の読み込み
        data = [line.strip("\n") for line in data]  #全ての行から"\n"を排除

    ans = []    #回答保存用配列

    #各行をループ
    for text in data:
        if text == "EOS":
            break
        dict = {}

        #空白の場合の処理
        if text[0] == " ":
            dict["surface"] = " "
            dict["base"] = " "
            temp = text.split("\t")
            print(temp)
            dict["pos"] =temp[0]
            dict["pos1"] = temp[1]
        #それ以外
        else:
            temp = text.split("\t")
            dict["surface"] = temp[0]
            dict["base"] = temp[1]
            dict["pos"] = temp[2]
            dict["pos1"] = temp[3]
        ans.append(dict)
    return ans
```

```python
import collections

datas = prepare()

# 単語の出現頻度計算
word_list = []
for data in datas:
    word_list.append(data["base"])

data = collections.Counter(word_list)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'MS Gothic'

sorted_data = sorted((data.values()), reverse = True)
print(sorted_data)
plt.plot(sorted_data)
plt.xlabel('出現頻度順位')
plt.ylabel('出現頻度')
ax = plt.gca()
ax.set_yscale('log')  # メイン: y軸をlogスケールで描く
ax.set_xscale('log')
plt.show()
```
