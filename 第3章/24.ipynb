```python
import json
import re
FILE = "./jawiki-country.json"

with open(FILE, 'r') as f:
    info_list = f.readlines()

json_list = []
for i in info_list:
    json_list.append(json.loads(i))

for row in json_list:
    if row["title"] == "イギリス":
        US_text = row["text"]

# urlの抽出(二次元の配列になっている箇所がある)
urls = []

US_list = US_text.split("\n")
for text in US_list:
    url = re.findall(r'http[^(\| | \s)]+', text) #httpで始まって空白で終わる箇所の抽出 => urlの抽出
    if url:
        urls.append(url)

# url_textをflattenしたリスト
url_list = sum(urls,[])
url_list
```
